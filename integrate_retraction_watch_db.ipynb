# Import and process data files

```python
cwd = Path.cwd()
db_file = cwd / '240304_retraction_watch_db.csv'

# with open('240304_retraction_watch_db.csv', 'r') as f:
#     a = f.readlines()

df = pd.read_csv('240304_retraction_watch_db.csv', encoding = 'latin')
df
```

```python
doi_list = list(df.OriginalPaperDOI)
```

```python
figure_params = [ {'folder': cwd / 'Case_crispr_cas9',
                   'year1': 2012, 'year2': 2023, 'min_size': 100, 'add_fit': False},
                  {'folder': cwd / 'Case_rnai_cancer',
                   'year1': 2002, 'year2': 2023, 'min_size': 70, 'add_fit': False},
                  {'folder': cwd / 'Case_circular_rna',
                   'year1': 2008, 'year2': 2023, 'min_size': 250, 'add_fit': False},
                  {'folder': cwd / 'Case_mirna_development',
                   'year1': 2004, 'year2': 2023, 'min_size': 400, 'add_fit': False},
                  {'folder': cwd / 'Case_mirna_cancer',
                   'year1': 2004, 'year2': 2023, 'min_size': 650, 'add_fit': False},
                  {'folder': cwd / 'Case_lncrna',
                   'year1': 2008, 'year2': 2023, 'min_size': 400, 'add_fit': False},
                ]
```

```python
count_retractions = 0
count_no_doi = 0
matched_dois = []
unmatched_retract_pubmed = []
for j, case in enumerate( figure_params[:1] ):
    case_folder = case['folder']
    folder = case_folder.parts[-1]
    print(f"-----{folder}-----\n")

    # Load data
    #
    articles_file = case_folder / 'articles_clean.json'
    with open(articles_file, 'r', encoding = 'utf-8') as f_json:
        articles = json.load(f_json)

    for article in articles:
        if article['retraction']:
            count_retractions += 1
            if article['doi'] in doi_list:
                matched_dois.append( doi_list.index(article['doi']) )
            else:
                if article['doi']:
                    unmatched_retract_pubmed.append( article['doi'] )
                else:
                    count_no_doi += 1

print( f"The DOI list from Retraction Watch has {len(doi_list)} records." )
print( f"We matched DOIs to {len(matched_dois)} of {count_retractions} " 
       f"retractions in PubMed.\nAnother {count_no_doi} retractions did not "
       f"have DOIs, and {len(unmatched_retract_pubmed)} did not match.")
```

```python
for j, case in enumerate( figure_params[:1] ):
    case_folder = case['folder']
    folder = case_folder.parts[-1]
    print(f"-----{folder}-----\n")

    # Load data
    #
    articles_file = case_folder / 'articles_clean.json'
    with open(articles_file, 'r', encoding = 'utf-8') as f_json:
        articles = json.load(f_json)
        
    articles_dois = []
    for article in articles:
        articles_dois.append(article['doi'])
        
    print(len(articles_dois))

    for i, doi in enumerate( doi_list ):
        if doi in articles_dois:
            print(i, doi)
```
